```python
# Does any column contain alpha and numeric data? Decide how to clean it.

pd.set_option("display.max_colwidth", 150)

def values_check(x,y):
    lst=[]
    for i in y:
        a = i
        b = x[i].count()
        if len(x[i].unique())>10: 
            c = '>10' 
        else: 
            c = []
            for j in range(0,len(x[i].value_counts())): 
                ab = x[i].value_counts().index[j] 
                ac = x[i].value_counts()[j]
                c.append((ab,ac))
        d = x.dtypes[i]
        lst.append((a,b,c,d))
    return pd.DataFrame(lst)

values_check(customer_df_obj,customer_df_obj.columns)

## columns policy and renew_offer_type contain both alpha and numeric data. 
## there is no need to clean renew_offer_type as the values are not ordinals.
## column policy will be reviewed separately
## also we have customer which values are alpha-numerical. This column could be just dropped. 
```

```python
# Check number of unique values in each column, can they be combined in any way to ease encoding? 
# Comment your thoughts and make those changes.

## discussed above

def edu_clean(x):
    if x in ['Bachelor','Master','Doctor']:
        x = 'Univercity'
    return x

customer_df_obj['education'] = customer_df_obj['education'].apply(edu_clean)

def emp_clean(x):
    if x in ['Unemployed','Medical Leave','Disabled','Retired']:
        x = 'Not Employed'
    return x

customer_df_obj['employmentstatus'] = customer_df_obj['employmentstatus'].apply(emp_clean)

def mari_clean(x):
    if x in ['Divorced']:
        x = 'Single'
    return x

customer_df_obj['marital_status'] = customer_df_obj['marital_status'].apply(mari_clean)

values_check(customer_df_obj,customer_df_obj.columns)

## I can not decide on 'vihicle_class': on one hand the total number of lux entries is small, and 
## lux_car and lux_suv can be joined into one. On the other hand categories 'car' (with 2 subcategories) 
## and 'suv' are separate categories in non-lux segment and we can loose useful information by merging them.
```
